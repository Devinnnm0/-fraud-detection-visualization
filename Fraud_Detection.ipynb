```python
# Install necessary library
!pip install openpyxl

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
df = pd.read_excel('/content/fraud_detection.xlsx')  # Adjust path if needed

# Quick cleanup if needed
df.columns = df.columns.str.strip()

# Display the first few rows to confirm
print(df.head())

# Grouped countplot: Merchant Type vs Fraud Status
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='merchant_type', hue='label', palette='viridis')

plt.title('Fraud vs Non-Fraud Transactions by Merchant Type')
plt.xlabel('Merchant Type')
plt.ylabel('Count')
plt.legend(title='Is Fraud', labels=['Non-Fraud', 'Fraud'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
